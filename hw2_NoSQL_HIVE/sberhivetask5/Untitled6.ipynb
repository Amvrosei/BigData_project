```python
data = pd.read_excel('analize_hw.xlsx')
data.columns
```

```python
data.plot(x = 'percent', y = ['Moscow', 'Leningrad', 'Krasnodar', 'Rostov', 'Sverdlovsk'], figsize=(9,6), marker='D')
plt.xlabel('percent')
plt.ylabel('counts of men+women')
plt.show()
```

```python
data.plot(x = 'Rows', y = ['Moscow_dev', 'Leningrad_dev', 'Krasnodar_dev', 'Rostov_dev', 'Sverdlovsk_dev'], figsize=(9,6))
plt.xlabel('Rows')
plt.ylabel('deviation')
plt.show()
```
